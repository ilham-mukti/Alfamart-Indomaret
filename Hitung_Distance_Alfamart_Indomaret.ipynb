```python
import pandas as pd
import haversine as hs
from haversine import Unit
import numpy as np

pd.options.display.max_columns = None
pd.options.display.max_rows = 20
```

```python
datas = pd.read_csv("/content/DATA_JAKSEL_FIXED_CLEANED.csv")
datas['nomor'] = datas.index
datas.head(3)
```

```python
datas_kecil_indomaret = datas[datas['store'] == "Indomaret"]
datas_kecil_alfamart = datas[datas['store'] == "Alfamart"]
datas_kecil_indomaret.shape, datas_kecil_alfamart.shape
```

```python
def convert(x, index):
  # Cari distance antar titik
  origin = (datas_kecil_alfamart.latitude[index], datas_kecil_alfamart.longitude[index])
  dist = (datas_kecil_indomaret.latitude[x], datas_kecil_indomaret.longitude[x])
  distance = hs.haversine(origin, dist, unit=Unit.METERS)
  return distance

for index in datas_kecil_alfamart['latitude'].index:
  datas_kecil_indomaret[datas_kecil_alfamart['latitude'][index]] = datas_kecil_indomaret.nomor.apply(convert, args=(index,))
```

```python
datas_kecil_indomaret
```

```python
# Cari titik mana yang paling dekat
datas_kecil_indomaret['jarak_terkecil_lat'] = datas_kecil_indomaret.iloc[:,13:].idxmin(axis=1)
datas = datas_kecil_indomaret.copy()
```

```python
datas.head(20)
```

```python
def convert_to_values(row):
  # Cari latitude yang paling deket berapa meter
  lat = row['jarak_terkecil_lat']
  return row[lat]

def convert_to_lng(row):
  return datas_kecil_alfamart[datas_kecil_alfamart['latitude'] == row['jarak_terkecil_lat']]['longitude'].values[0]

def convert_to_alamat(row):
  return datas_kecil_alfamart[datas_kecil_alfamart['latitude'] == row['jarak_terkecil_lat']]['alamat_tempat'].values[0]

def convert_to_nama(row):
  return datas_kecil_alfamart[datas_kecil_alfamart['latitude'] == row['jarak_terkecil_lat']]['nama_tempat'].values[0]

datas['jarak_terkecil_lng'] = datas.apply(convert_to_lng, axis=1)
datas['jarak_terkecil_nama_tempat'] = datas.apply(convert_to_nama, axis=1)
datas['jarak_terkecil_alamat'] = datas.apply(convert_to_alamat, axis=1)
datas['jarak_terkecil_values'] = datas.apply(convert_to_values, axis=1)
```

```python
def convert_to_yes_no(x):
  # Jika jaraknya dibawah 30 meter, berarti dekat
  if(x <= 50):
    return "yes"
  else:
    return "no"
```

```python
datas['berdekatan_30m'] = datas['jarak_terkecil_values'].apply(convert_to_yes_no)
```

```python
datas['berdekatan_50m'] = datas['jarak_terkecil_values'].apply(convert_to_yes_no)
```

```python
datas_save = pd.concat([datas.iloc[:,0:13], datas.iloc[:,-7:]], axis=1)
datas_save
```

```python
datas_save.to_csv("data_final_alfamart_indomaret.csv", index=False)
```

```python
datas = pd.read_csv("/content/DATA_JAKSEL_FIXED_CLEANED.csv")
datas_fixed = pd.read_csv("/content/data_final_alfamart_indomaret.csv")
datas.head(3)
```

```python
datas_berdekatan = datas_fixed[datas_fixed['berdekatan_50m'] == "yes"]
datas_berdekatan.head(3)
```

```python
def tes_convert(row):
  berdekatan_indomaret_list = datas_berdekatan['latitude'].tolist()
  berdekatan_alfamart_list = datas_berdekatan['jarak_terkecil_lat'].tolist()
  x = "no"
  if row in berdekatan_indomaret_list:
    x = "yes"

  if row in berdekatan_alfamart_list:
    x = "yes"
  return x

datas['berdekatan_50m'] = datas.latitude.apply(tes_convert)
```
